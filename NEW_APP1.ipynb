# Property Tax Rate CSV

The Levy Year 2012 real property tax rate dataset reflects all the rates per $100 set each year by the County Council. These rates are applied to the assessed value of the real property to derive the applicable ad valorem tax. The fully phased-in assessed value equals the full cash value of the property. The real property tax is levied annually on all taxable land and improvements. The tax rate schedule is updated annually. A detailed tax rate schedule for all tax classes is available by clicking on “Tax Rates” under the link for “County Taxes” when visiting the County website at www.montgomerycountymd.gov/finance or by calling 311 (240-777-0311).

https://catalog.data.gov/dataset/real-property-tax-rates

The business questions I am answering in this dataset are:
1. What tax class has the lowest total tax rate?
2. which tax class has the lowest TOTAL SPECIAL SERVICE AREA TAX?

## What tax class has the lowest total tax rate?

The lowest tax rate is useful because it's optimal for saving money.

```python
lowest_tax= tax_data
lowest_tax= lowest_tax.loc[lowest_tax['TOTAL TAX RATE'] == lowest_tax['TOTAL TAX RATE'].min()]
lowest_tax.head()
```

## which tax class has the lowest TOTAL SPECIAL SERVICE AREA TAX?

This is the area that has the least amount of tax

```python
# Recreating dataset so I have NaN values again
tax_data2 = pd.read_csv("Real_Property_Tax_Rates.csv")
tax_data2.rename(columns={'TAX CLASS ': 'TAX_CLASS'}, inplace=True)
#tax_data2.set_index("TAX_CLASS", inplace=True)
tax_data2.head(10)
```

```python
tax_data2.fillna(0, inplace=True)
```

```python
tax_data2.plot(x='TAX_CLASS' ,y='TOTAL SPECIAL SERVICE AREA TAX', kind='scatter', figsize=(20,10), rot=100)
```

```python
# tax classes with the lowest TOTAL SPECIAL SERVICE AREA TAX
lowest_tax= tax_data2[['TAX_CLASS','TOTAL SPECIAL SERVICE AREA TAX']]
lowest_tax= lowest_tax.loc[lowest_tax['TOTAL SPECIAL SERVICE AREA TAX'] == lowest_tax['TOTAL SPECIAL SERVICE AREA TAX'].min()]
lowest_tax.head()

```

```python
# alternatively I could find the tax classes with the lowest TOTAL SPECIAL SERVICE AREA TAX
# by doing this
alternate = tax_data2[['TAX_CLASS','TOTAL SPECIAL SERVICE AREA TAX']].sort_values(by='TOTAL SPECIAL SERVICE AREA TAX')
alternate.head(10)
```
